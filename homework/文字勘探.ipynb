```python
ll=open('D:\\345.txt','r')
l=ll.read()
```

```python
pp=open('D:\\123.txt','r')
p=pp.read()
```

```python
dic={}
for ele in jieba.cut(l):
    if ele not in dic:
        dic[ele]=1
    else:
        dic[ele]=1+dic[ele]
```
